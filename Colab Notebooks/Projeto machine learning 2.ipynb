```
import pandas as pd

uri = "https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv"
dados = pd.read_csv(uri)
dados.head()
```

```
x = dados[["home","how_it_works","contact"]]
y = dados[["bought"]]
print(y)
```

```
#Treino com 75 elementos e teste com 24 elementos
treino_x = x[:75]
treino_y = y[:75]
teste_x = x[75:]
teste_y = y[75:]
```

```
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

modelo = LinearSVC() #Criando modelo linear
modelo.fit(treino_x, treino_y) #Adaptando modelo aos treinos
previsoes = modelo.predict(teste_x) #Previsoes do computador

precisao= accuracy_score(teste_y, previsoes) * 100
print("A precisao foi de %.2f%%" %precisao)
```
